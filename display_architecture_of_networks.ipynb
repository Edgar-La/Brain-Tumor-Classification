# Import modules

```python
import os
import visualkeras
import tensorflow as tf
from tensorflow.keras.utils import plot_model
from tensorflow.keras.applications import VGG16, ResNet50V2, InceptionResNetV2, MobileNetV2, EfficientNetB3
```

# Get names of the Networks used

```python
path_to_models = 'Saved_models_and_history'
filenames = [filename[:-13] for filename in os.listdir(path_to_models) if filename.endswith('.json')]
filenames
```

```python
visualkeras.layered_view(VGG16(weights='imagenet', include_top = False, input_shape=(256,256,3)),
                         legend = True, to_file='Architectures/'+filenames[0][:-9]+'.png')
```

```python
i = 0
print(filenames[i])
visualkeras.layered_view(tf.keras.models.load_model(path_to_models+'/'+filenames[i]), legend = True, to_file='Architectures/'+filenames[i]+'.png')
```

```python
i = 1
print(filenames[i])
visualkeras.layered_view(tf.keras.models.load_model(path_to_models+'/'+filenames[i]), legend = True, to_file='Architectures/'+filenames[i]+'.png')
```

```python
plot_model(tf.keras.models.load_model(path_to_models+'/'+filenames[i]),
                                      to_file= 'Architectures/model_plot_'+filenames[i]+'.png',
                                      show_shapes=True,
                                      show_layer_names=True)
```
